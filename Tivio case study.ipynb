## a.	How many married men who are early adopters have monthly electronics spend high enough that they can afford to purchase a TiVo for $499 and still be able to spend more on electronics in the next two years? 

```python
# Filtering out married man who were early adopters

early_adopters = tivio_data[(tivio_data['Gender'] == 'male') & (tivio_data['Marital Status'] == 'married')&(tivio_data['Technology Adoption'] == 'early')]

tivo_cost = 499

affordability_condition = (tivio_data['Monthly Electronics Spend'] * 24) > tivio_cost + (tivio_data['Monthly Electronics Spend'] * 12 * 2)


able_to_purchase_tivo = early_adopters[early_adopters['Monthly Electronics Spend'] * 24 > tivo_cost]
num_married_men_able_to_purchase_tivo = len(able_to_purchase_tivo)
num_married_men_able_to_purchase_tivo
```

## d. How many seniors (over the age of 65) spend more than six hours a day watching TV? What is their income range? What is their average annual income? 

```python
seniors_watching_tv = tivio_data[(tivio_data['Age'] > 65) & (tivio_data['TV Viewing (hours/day)'] > 6)]
num_seniors_watching_tv = len(seniors_watching_tv)
income_range_seniors_watching_tv = seniors_watching_tv['Annual Income (x1000 $)'].min(), seniors_watching_tv['Annual Income (x1000 $)'].max()
average_income_seniors_watching_tv = seniors_watching_tv['Annual Income (x1000 $)'].mean()
num_seniors_watching_tv, income_range_seniors_watching_tv, average_income_seniors_watching_tv
```

```python
seniors_wtachingTV_six_hours_a_day = tivio_data[(tivio_data['Age'] > 65) & (tivio_data['TV Viewing (hours/day)'] > 6)]

count_active_seniors = len(seniors_wtachingTV_six_hours_a_day)
seniors_income_range = seniors_wtachingTV_six_hours_a_day['Annual Income (x1000 $)'].min(), seniors_wtachingTV_six_hours_a_day['Annual Income (x1000 $)'].max()
average_income_seniors = seniors_wtachingTV_six_hours_a_day['Annual Income (x1000 $)'].mean()

print(f"Number of active seniors: {num_active_seniors}")
print(f"Income range of seniors: {seniors_income_range}")
print(f"Average income of active seniors: {average_income_seniors}")
```
